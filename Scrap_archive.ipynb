```python
from bs4 import BeautifulSoup
from requests.adapters import HTTPAdapter
import requests
```

```python
URL_dollar = 'https://www.tgju.org/profile/price_dollar_rl/history'
URL_euro = 'https://www.tgju.org/profile/price_eur/history'
URL_AED = 'https://www.tgju.org/profile/price_aed/history'
```

```python
URL = 'https://www.tgju.org/profile/price_dollar_rl/history'
session = requests.Session()
retry = HTTPAdapter(max_retries=5)
session.mount('https://', retry)
html_text = session.get(URL).text
soup=BeautifulSoup(html_text,'lxml')
```

```python
default_table = soup.find('div',class_='tables-data table-mobile-grid')
```

```python
rows = default_table.find_all('tr')
```

```python

for row in rows:
    elements = row.find_all('td')
    i = 0
    for element in elements:
        if i == 0:
            reopening_rate = element.text
            print(f'Reopening rate : {reopening_rate}')
        elif i == 1:
            lowest_rate = element.text
            print(f'lowest rate : {lowest_rate}')
        elif i == 2:
            highest_rate = element.text
            print(f'highest rate : {highest_rate}')
        elif i == 3:
            final_rate = element.text
            print(f'final rate : {final_rate}')
        elif i ==4:
            rate_change = element.text
            print(f'rate of change : {rate_change}')
        elif i == 5:
            percent_change = element.text
            percent_change = percent_change.replace('\n','')
            print(f'percent of change : {percent_change}')
        elif i == 6:
            AD_date = element.text
            print(f'AD data : {AD_date}')
        elif i ==7:
            solar_date = element.text
            print(f'solar date : {solar_date}')
        i +=1
    print('--------------------------------------------')
```
